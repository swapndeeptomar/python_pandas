```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
b = np.array((5, 6, 7, 8))
c = np.zeros((2, 3))
d = np.ones((3, 3))
e = np.eye(3)
f = np.arange(0, 10, 2)
g = np.linspace(0, 1, 5)

h = np.random.rand(3, 3)
i = np.random.randint(1, 10, (2, 2))

print(a, b, c, d, e, f, g)
```

```python
print(a.shape, c.shape)
print(a.ndim, c.ndim)
print(a.size, c.size)
print(a.dtype)
a = a.astype(float)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x ** 2)
```

```python
m = np.array([[1, 2], [3, 4]])
n = np.array([[5, 6], [7, 8]])

print(m + n)
print(m @ n)
print(np.dot(m, n))
```

```python
print(np.sum(m))
print(np.mean(m))
print(np.std(m))
print(np.min(m))
print(np.max(m))

print(np.sum(m, axis=0))
print(np.sum(m, axis=1))
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])
print(arr[-1])
print(arr[1:4])
print(arr[::-1])
```

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])

print(mat[0, 1])
print(mat[:, 1])
print(mat[1, :])
```

```python
reshaped = np.arange(12).reshape(3, 4)
print(reshaped)

flattened = reshaped.flatten()
print(flattened)

ravelled = reshaped.ravel()
print(ravelled)
```

```python
stack1 = np.array([1, 2])
stack2 = np.array([3, 4])

print(np.vstack((stack1, stack2)))
print(np.hstack((stack1, stack2)))
```

```python
split_arr = np.array([1, 2, 3, 4, 5, 6])
print(np.split(split_arr, 3))

broadcast_a = np.array([1, 2, 3])
broadcast_b = 2

print(broadcast_a + broadcast_b)
```

```python
cond = np.array([1, 2, 3, 4, 5])
print(cond[cond > 2])

print(np.where(cond > 3, 1, 0))
```

```python
rand_arr = np.random.randint(1, 100, 10)

print(np.sort(rand_arr))
print(np.unique(rand_arr))

print(np.argmax(rand_arr))
print(np.argmin(rand_arr))

print(np.cumsum(rand_arr))
print(np.cumprod(rand_arr))
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.concatenate((a, b)))
```

```python
matrix = np.array([[1, 2], [3, 4]])

print(np.transpose(matrix))

inv = np.linalg.inv(matrix)
print(inv)

det = np.linalg.det(matrix)
print(det)

eigvals, eigvecs = np.linalg.eig(matrix)
print(eigvals)
print(eigvecs)
```

```python
data = np.array([1, 2, 3, 4, 5])

print(np.percentile(data, 50))
print(np.median(data))

print(np.clip(data, 2, 4))

print(np.log(data))
print(np.exp(data))
print(np.sqrt(data))
```
